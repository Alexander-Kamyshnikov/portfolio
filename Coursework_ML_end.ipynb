### Предварительное обучение модели на базовых данных

создадим функцию, которая на вход будет принимать список обученных моделей, а на выходе таблица со значениями 'roc_auc_score'

```python
def roc_auc_df(models, name_models, test, targets):


  """функция принимает на вход:
    -  список моделей - models,
    -  cписок наименований моделей - name_models,
    -  датасет признаков - test,
    - датасет целевых признаков - targets
    Возвращает датафрейм Pandas с оценкой roc_auc для каждой модели,
    датафрейм отсортирован по убыванию оценок
  """

  roc_auc_list = []
  for clf in models:
    
    # для модели xgb необходимо предварительно создать матрицу признаков и отдельно посчитать predict
    # тоже самое для моделей линейных моделей
      
    if str(type(clf)) == "<class 'xgboost.core.Booster'>":
      xgb_features = xgb.DMatrix(test, targets)
      pred_proba = clf.predict(xgb_features) 
    elif str(type(clf)) == "<class 'sklearn.linear_model._base.LinearRegression'>":
      pred_proba = clf.predict(test)
    elif str(type(clf)) == "<class 'sklearn.linear_model._ridge.Ridge'>":
      pred_proba = clf.predict(test)
    else:    
      pred_proba = clf.predict_proba(test)[:,1]
    # scor =  roc_auc_score(targets, 
    #           pred_proba)                                
    roc_auc_list.append(roc_auc_score(targets, pred_proba))
    
  df = pd.DataFrame( {'roc_auc': roc_auc_list}, index=name_models)
  df = df.sort_values('roc_auc',ascending=False)
  return df
    
    # [elem for elem in '[ gbc, dtc, rfc, cat.....]'.strip('[]').split(', ')]

```

Для дальнейшего преобразования категориальных данных в значения средних от таргета, перобразуем непреобразованные категориальные данные обратно в тип 'object'

Для каждого категориального признака мы имеем несколько значений от 1 до 4х. Необходимо найти количество каждого значения для целевой переменной равной 1. Создадим соответствующую функцию для вычисления и присвоения значениям признаков значений target mean

```python
def mean_target_enc(X_train, y_train, x_test, skf):   
    
    train_df = X_train
    valid_df = x_test
    
    glob_mean = y_train.mean()
    train_df = pd.concat([train_df, pd.Series(y_train, name='Churn')], axis=1)
    new_train_df = train_df.copy()
    
    cat_features = train_df.columns[train_df.dtypes == 'object'].tolist()    

    for col in cat_features:
        new_train_df[col + '_mean_target'] = [glob_mean for _ in range(new_train_df.shape[0])]

    for train_idx, valid_idx in skf.split(train_df, y_train):
        train_df_cv, valid_df_cv = train_df.iloc[train_idx, :], train_df.iloc[valid_idx, :]

        for col in cat_features:
            
            means = valid_df_cv[col].map(train_df_cv.groupby(col)['Churn'].mean())
            valid_df_cv[col + '_mean_target'] = means.fillna(glob_mean)
            
        new_train_df.iloc[valid_idx] = valid_df_cv
    
    new_train_df.drop(cat_features + ['Churn'], axis=1, inplace=True)
    
    for col in cat_features:
        means = valid_df[col].map(train_df.groupby(col)['Churn'].mean())
        valid_df[col + '_mean_target'] = means.fillna(glob_mean)
        
    valid_df.drop(train_df.columns[train_df.dtypes == 'object'], axis=1, inplace=True)
    
    return new_train_df, valid_df
```

```python
dtrain = xgb.DMatrix(mean_X_train, y_train)
dtest = xgb.DMatrix(mean_x_test, y_test)
num_round = 30
```

Для поиска оптимальных гиперпараметров модели xgboost воспользуемся библиотекой hyperopt. Библиотека позволяет строить математическую гипотезу, как должна выглядеть результирующая функция по переданным параметрам и проверяет эту гипотезу иттеративно. Создадим функции для реализации подбора параметров

```python
from hyperopt import hp, tpe, fmin, STATUS_OK, Trials
```

```python
def score(params):
    from sklearn.metrics import log_loss
    params['max_depth'] = int(params['max_depth'])
    dtrain = xgb.DMatrix(mean_X_train, label=y_train)
    dvalid = xgb.DMatrix(mean_x_test, label=y_test)
    model = xgb.train(params, dtrain)
    predictions = model.predict(dvalid).reshape((x_test.shape[0], 2))
    score = log_loss(y_test, predictions)

    return {'loss': score, 'status': STATUS_OK}
```

```python
def optimize(trials):
    space = {

             'learning_rate': hp.quniform('eta', 0.005, 0.05, 0.005),
             'max_depth': hp.quniform('max_depth', 2, 42, 1),
             'min_child_weight': hp.quniform('min_child_weight', 10, 100, 10),
             'subsample': hp.quniform('subsample', 0.5, 1, 0.05),
             'gamma': hp.quniform('gamma', 0.5, 1, 0.01),
             'colsample_bytree': hp.quniform('colsample_bytree', 0.4, 1, 0.05),
             'num_class' : 2,
             'eval_metric': 'aucpr',
             'objective': 'multi:softprob',
             'nthread' : 4

             }
    
    best = fmin(score, space, algo=tpe.suggest, trials=trials, max_evals=10)
    return best
```

Так как hyperopt выдаёт различные вариации параметров, перебором найдём вариант с самым высоким скором

```python
def xgb_best_score(n=5):
    
    options = []
    list_params =[]
    
    for x in range(n):
        trials = Trials()
        best_params = optimize(trials)
        options.append(best_params)

    for y in range(len(options)):   
        dict_params ={}
        dict_params.setdefault('colsample_bytree', round(options[y].get('colsample_bytree'),2))
        dict_params.setdefault('eta', options[y].get('eta'))
        dict_params.setdefault('gamma', round(options[y].get('gamma'),2))
        dict_params.setdefault('max_depth', int(options[y].get('max_depth')))
        dict_params.setdefault('min_child_weight', int(options[y].get('min_child_weight')))
        dict_params.setdefault('subsample', round(options[y].get('subsample'),2))
        list_params.append(dict_params)

    best_roc = []
    for param in list_params:
        xgb_tt = xgb.train(param, dtrain, num_round)
        predict = xgb_tt.predict(dtest)
        best_roc.append(roc_auc_score(y_test, predict))

    list_params = np.asarray(list_params)
    best_roc = np.asarray(best_roc)   

    best_param = list_params[np.where(best_roc == best_roc.max())[0][0]]
    
    print('best roc_auc: {}\nbest params:{}'.format(best_roc.max(), best_param))
```

### 4.1 Создание стэка из предиктов обученных моделей

На данном этапе мы подготовили и преобразовали данные, обучили и настроили 7 моделей. Показатель roc_auc  у лучших моделей выше 0,77. У худших моделей он ниже 0,75 но при этом высокий показатель roc_auc, а он в конечном итоге нас интересует. Для того что бы ещё повысить показатели метрики, попробуем объеденить предикты моделей в общий датафрейм и уже на его основе построить мета модель. Будем ожидать повышение метрик. Для этого создадим функцию, которая будет принимать настроенные модели и данные для обучения,  предикты будем записывать в матрицу и эти предикты будут новыми признаками. Мета модель будем обучать на новых мета признаках. 

```python
def stack_predict(models, X_train, y_train, cv=5, random_state=rs):
  mtrx = np.empty((X_train.shape[0], len(models)))
  for n, model in enumerate(models):
  # для модели xgb необходимо предварительно создать матрицу признаков и отдельно посчитать predict

    if str(type(model)) == "<class 'xgboost.core.Booster'>":
        
        xgb_features = xgb.DMatrix(X_train, y_train)
        mtrx[:,n] = model.predict(xgb_features) 
    elif str(type(model)) == "<class 'sklearn.linear_model._ridge.Ridge'>":
        mtrx[:,n] = model.predict(X_train)
    else:    
        mtrx[:,n] = model.predict_proba(X_train)[:,1]
  mtrx_pd = pd.DataFrame(mtrx)
  return(mtrx_pd)
```

### 5.2 Визуализация полученных моделей

```python
def display_scores(model,features, target, proba=0.5):
    
    """функция считающая метрики: precission, recall, roc_auc_score. Выводит на экран: 
       две круговые диаграммы  отношений TP к FP и TP к FN  как precision и recall, 
       график ROC AUC со значением этой метрики. Также два графика, первый для наглядности
       отношения всех таргетов possitive  к таргетам negative, под ним с соблюдением пропорций
       значения TP FN FP TN. Такой график позволяет наглядно видеть сколько 'лишних' объектов
       посчитала модель за позитивных и сколько неправильно определила как негативные"""
    
    
    
    # определение основных метрик
    # задание порога вероятности
    
    # Для xgboost необходимо отдельно создать predict вероятностей
    if str(type(model)) == "<class 'xgboost.core.Booster'>":
        pred_proba = model.predict(features)
    else:    
        
        pred_proba = model.predict_proba(features)[:,1]
        
    proba_mod = np.where(pred_proba>proba,1,0)
    conf_matrix = confusion_matrix(target, proba_mod)
    precision = precision_score(target, proba_mod)
    recall = recall_score(target, proba_mod)
    roc_auc = roc_auc_score(target, proba_mod)
    
    
    #  ввывод на печать наименования модели
    print(type(model).__name__)
    
    # отрисовка полученных значений на графике
    
    
    
    fpr, tpr, threshold = roc_curve(target, proba_mod)
    roc_auc = roc_auc_score(target,proba_mod)

    fig = plt.figure(figsize=(20,10))


    def func(pct,allvals):
        absolute = int(pct/100.*np.sum(allvals))
        return "{:.1f}%\n ({:d} )".format(pct, absolute)

    ax_1 = fig.add_subplot(2, 3, 1)
    ax_2 = fig.add_subplot(2, 3, 2)
    ax_3 = fig.add_subplot(2, 3, 3)
    ax_4 = fig.add_subplot(10, 1, 7)
    ax_5 = fig.add_subplot(10, 1, 8)
    
    
    ax_1.pie([ conf_matrix[1][1],  conf_matrix[0][1]], 
             textprops=dict(color="black", fontsize=16),
             startangle=140,
             shadow=True,
             autopct= lambda pct: func(pct, [ conf_matrix[1][1],  conf_matrix[0][1]]),
             labels = ['TP', 'FP']
            )

    ax_2.pie([ conf_matrix[1][1],  conf_matrix[1][0]], 
             textprops=dict(color="black", fontsize=16),
             startangle=140,
             shadow=True,
             autopct= lambda pct: func(pct, [ conf_matrix[1][1],  conf_matrix[0][1]]),
             labels = ['TP', 'FN']
            )
    
    
    inc = 0
    pn = ['TP', 'FN', 'FP', 'TN']
    pos_neg = ['Positive', 'Negative']
    colors = ['green', 'lime', 'yellow', 'orange'] 
    tp = np.unique(target, return_counts=True)

    for i, x in enumerate(tp[1][::-1]):
        ax_4.bar(x/2 + inc, 1, x, color=colors[i*3] )
        ax_4.text(x/2 + inc-150, 0.3, pos_neg[i], fontsize=24 )
        inc += x

    inc=0    
    for i, x in enumerate(conf_matrix.reshape(-1)[::-1]):
        ax_5.bar(x/2 + inc, 1, x, color=colors[i])
        ax_5.text(x/2 + (inc-35), 0.3, pn[i], fontsize=20 )
        ax_5.text(x/2 + inc, 0.6, str(x), fontsize=12)
        inc += x

    ax_4.axis('off')
    ax_5.axis('off')    
    ax_3.plot(fpr, tpr, 'b')
    ax_3.set_ylabel('True Positive Rate')
    ax_3.set_xlabel('False Positive Rate')
    ax_1.set(title = 'precission = '+ str(round(precision,6)), xticks=[], yticks=[])
    ax_2.set(title = 'recall = ' + str(round(recall,6)), xticks=[], yticks=[])
    ax_3.set(title = 'ROC AUC = ' + str(round(roc_auc,6)), xticks=[], yticks=[])

    plt.show()
```

Подберём для каждой лучшей модели порог вероятности соотнесения к классу таким образом, что бы в каждой модели оказалось одинаковое количество неправильно ложно негативных предиктов, тогда по количеству ложно позитивных можно выбрать лучшую модель
